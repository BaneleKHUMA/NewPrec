```python
# Hard-coded list of 'Yield_tons_per_hectare' values
yield_data = [6.55, 8.52, 1.12, 6.51, 7.24, 2.93, 5.76, 1.34, 7.31, 2.07]

# Function for Linear Search
def linear_search(arr, target):
    for index, value in enumerate(arr):
        if value == target:
            return index  # Return the index of the found target
    return -1  # Return -1 if the target is not found

# Function for Binary Search (on a sorted list)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid  # Return the index if found
        elif arr[mid] < target:
            left = mid + 1  # Search the right half
        else:
            right = mid - 1  # Search the left half
    return -1  # Return -1 if the target is not found

# --- Linear Search ---
target_value = 6.554 # Replace with the yield value you want to search

# Run Linear Search on the hard-coded yield_data
linear_search_result = linear_search(yield_data, target_value)
if linear_search_result != -1:
    print(f"Linear Search: Found {target_value} at index {linear_search_result} in the unsorted data.")
else:
    print(f"Linear Search: {target_value} not found in the unsorted data.")

# --- Binary Search ---
# Sort the yield_data for binary search
sorted_yield_data = sorted(yield_data)

# Run Binary Search on the sorted yield_data
binary_search_result = binary_search(sorted_yield_data, target_value)
if binary_search_result != -1:
    print(f"Binary Search: Found {target_value} at index {binary_search_result} in the sorted data.")
else:
    print(f"Binary Search: {target_value} not found in the sorted data.")

```
