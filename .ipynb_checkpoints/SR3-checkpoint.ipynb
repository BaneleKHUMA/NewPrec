```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv(r"C:\Users\Banesta\Downloads\archive\crop_yield.csv")
```

```python
# Calculate the average yield per region
avg_yield_per_region = data.groupby('Region')['Yield_tons_per_hectare'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Yield_tons_per_hectare', data=avg_yield_per_region)
plt.title('Average Crop Yield per Region')
plt.xlabel('Region')
plt.ylabel('Average Yield (tons per hectare)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average yield and rainfall
avg_yield_rainfall = data.groupby('Rainfall_mm')['Yield_tons_per_hectare'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Rainfall_mm', y='Yield_tons_per_hectare', data=avg_yield_rainfall)
plt.title('Average Crop Yield vs Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Average Yield (tons per hectare)')
plt.grid()
plt.show()
```

```python
# Create a scatter plot with different colors for each crop type
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Temperature_Celsius', y='Yield_tons_per_hectare', hue='Crop', data=data, palette='Set2', alpha=0.7)

# Customizing the plot
plt.title('Crop Yield vs Temperature by Crop Type')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Yield (tons per hectare)')
plt.legend(title='Crop Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Create a pair plot for selected numerical features
selected_columns = ['Rainfall_mm', 'Temperature_Celsius', 'Yield_tons_per_hectare']
sns.pairplot(data[selected_columns])
plt.suptitle('Pair Plot of Selected Features', y=1.02)  # Adjust title position
plt.show()
```

```python
# Calculate summary statistics for numeric columns
summary_statistics = data[['Rainfall_mm', 'Temperature_Celsius', 'Yield_tons_per_hectare']].describe()

# Print the summary statistics
print(summary_statistics)
```

```python
# Calculate mean
mean_values = data[['Rainfall_mm', 'Temperature_Celsius', 'Yield_tons_per_hectare']].mean()

# Calculate median
median_values = data[['Rainfall_mm', 'Temperature_Celsius', 'Yield_tons_per_hectare']].median()

# Calculate standard deviation
std_dev_values = data[['Rainfall_mm', 'Temperature_Celsius', 'Yield_tons_per_hectare']].std()

# Print the statistics
print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nStandard Deviation:\n", std_dev_values)
```
